```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

df = pd.read_csv('C:\\Users\\ajroy\\OneDrive\\Desktop\\New folder\\loan_apporaval_model\\loan_prediction.csv')
print(df.head())
```

```python
df.isnull().sum()
```

```python
#dealing with  gender column missing values
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Gender'].unique()
df['Gender'].isnull().sum()
```

```python
#dealing with Marital Status column missing values
df['Married'].fillna(df['Married'].mode()[0], inplace=True)
df['Married'].unique()
df['Married'].isnull().sum()
```

```python
#dealing with dependents column missing values
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace=True)
df['Dependents'].unique()
df['Dependents'].isnull().sum()
```

```python
#dealing with self employed column missing values
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0], inplace=True)
df['Self_Employed'].unique()
df['Self_Employed'].isnull().sum()
```

```python
df.fillna({
    'Loan_Amount_Term': df['Loan_Amount_Term'].median(),
    'LoanAmount': df['LoanAmount'].median()
}, inplace=True)
```

```python
#dealing with Credit_History column missing values
df['Credit_History'].fillna(df['Credit_History'].mode()[0], inplace=True)
df['Credit_History'].isnull().sum()
```

```python
import pandas as pd

# Assume 'data' is your cleaned DataFrame
# Example: data.head() dikhata hai clean result

# Save it as a new CSV file
df.to_csv("loan_cleaned.csv", index=False)

# Verify if saved correctly
print("âœ… Cleaned file saved successfully as 'loan_cleaned.csv'!")

# Optional: Read again to double-check
check = pd.read_csv("loan_cleaned.csv")
print("ðŸ“„ New file shape:", check.shape)
print(check.head())

```
