```python
#vislulization all coulmns in the dataframe 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('C:\\Users\\ajroy\\OneDrive\\Desktop\\New folder\\loan_apporaval_model\\loan_prediction.csv')
df.info()
```

```python
# Visualizing all columns in the dataframe
df.hist(bins=25, figsize=(10,15))
plt.show()
```

```python

df['Loan_Status'].value_counts().plot(kind='bar')
plt.show()
```

```python
sns.lmplot(x='ApplicantIncome', y='LoanAmount', data=df, hue='Loan_Status', aspect=1.5)
plt.title('Income vs Loan Amount (with Trend Line)')
plt.show()
```

###  Insights from the Plot: *Applicant Income vs Loan Amount (with Trend Line)*

1. **Positive Relationship:**  
   As *ApplicantIncome* increases, *LoanAmount* also tends to increase. This shows that applicants with higher income usually apply for (or are eligible for) higher loan amounts.

2. **Loan Approval Pattern:**  
   The blue trend line (Loan_Status = Y) is steeper than the orange one (Loan_Status = N), suggesting that higher-income applicants are **more likely to get loan approval**.

3. **Concentration at Lower Income:**  
   Most applicants fall within the lower income range (below 10,000), indicating that the dataset is dominated by low-income groups.

4. **Presence of Outliers:**  
   A few data points show very high income (40,00080,000), which could represent rare cases or potential outliers worth separate analysis.

5. **Model Insight:**  
   The pattern confirms that *ApplicantIncome* has a strong impact on *LoanAmount* and approval decisions  making it an important feature for model prediction.
